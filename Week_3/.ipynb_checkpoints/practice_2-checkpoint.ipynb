```python
import tkinter as tk

# Handling button click event
def button_click():
    print("Button clicked!")

    # Show an information message box
    msgbox.showinfo("Info", "Welcome to COS 102 GUI App!\n")

    # Ask for user confirmation
    result = msgbox.askyesno("Confirmation", "Do you wanna continue?")

# Create the main window
root = tk.Tk()
root.title("Home Page")
root.geometry("300x100")

# Add a label widget
label = tk.Label(root, text="Hello Friend! \n")
label.pack()

# Add a button widget
button = tk.Button(root, text="Click Me!", command=button_click)
button.pack()

#Styling the button widget
button.config(fg="red", bg="yellow")

# Start the event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Login")
root.geometry("300x200")

def welcomeMessage(username):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("300x150")

    label_1 = tk.Label(window, text=f"Welcome {username}")
    label_1.pack(pady=10)

    label_2 = tk.Label(window, text="This is Python GUI with Tkinter")
    label_2.pack()

def submit():
    username = username_entry.get()
    password = password_entry.get()

    if username == "Mary" and password == "cos102":
        welcomeMessage(username)
    else:
        messagebox.showerror("Login", "Invalid username/password")

# UI
tk.Label(root, text="Username").pack()
username_entry = tk.Entry(root)
username_entry.pack()

tk.Label(root, text="Password").pack()
password_entry = tk.Entry(root, show="*")
password_entry.pack()

tk.Button(root, text="Login", command=submit).pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
import pandas as pd

# Load CSV
df = pd.read_csv("GIG-logistics.csv")
df["Name"] = df["Name"].str.lower().str.strip()
df["Department"] = df["Department"].str.lower().str.strip()

# Function to check employee
def check_employee():
    name = name_entry.get().lower().strip()
    dept = dept_entry.get().lower().strip()

    result.delete("1.0", tk.END)

    if (df["Name"] == name).any() and (df["Department"] == dept).any():
        result.insert(tk.END, f"Welcome {name.title()}!\n Here are other members of {dept.title()}: \n\n")

        members = df[df["Department"] == dept]["Name"]
        for m in members:
            result.insert(tk.END, m.title() + "\n")
    else:
        result.insert(tk.END, "Employee not found")

# GUI
root = tk.Tk()
root.title("Employee Checker")
root.configure(bg="#ffe6f0")
root.config(padx=20, pady=20)

# Heading
tk.Label(root, text="GIG Employee Checker",
         font=("Helvetica", 18, "bold"),
         bg="#ffe6f0", fg="#cc0066").pack(pady=10)

tk.Label(root, text="Name", bg="#ffe6f0").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department", bg="#ffe6f0").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Check 💖", command=check_employee, bg="#ff66a3").pack(pady=10)

result = tk.Text(root, height=10)
result.pack()

root.mainloop()
```
