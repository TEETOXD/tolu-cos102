# STORE CHECKOUT PSEUDOCODE
START

INPUT total_cost
OUTPUT "Total amount =", total_cost
REPEAT
    INPUT payment_amount
    IF payment_amount ≥ total_cost THEN
        change = payment_amount − total_cost
        OUTPUT "Payment accepted"
        OUTPUT "Change =", change
        OUTPUT "Thank you for shopping"
        EXIT LOOP
    ELSE
        OUTPUT "Insufficient payment"
        OUTPUT "Please provide additional payment"
    ENDIF
UNTIL payment_amount ≥ total_cost

STOP

```python
# CUBIC, QUARTIC, QUADRATIC EQUATION ROOTS CODE
import cmath  # For complex numbers
import numpy as np  # For cubic and quartic roots

def find_quadratic_roots(a, b, c):
    # Quadratic formula: ax^2 + bx + c = 0
    delta = cmath.sqrt(b**2 - 4*a*c)
    root1 = (-b + delta) / (2*a)
    root2 = (-b - delta) / (2*a)
    return root1, root2

def find_cubic_roots(a, b, c, d):
    # Cubic: ax^3 + bx^2 + cx + d = 0
    coefficients = [a, b, c, d]
    roots = np.roots(coefficients)
    return roots

def find_quartic_roots(a, b, c, d, e):
    # Quartic: ax^4 + bx^3 + cx^2 + dx + e = 0
    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)
    return roots

def main():
    print("Select the type of equation to solve:")
    print("1. Quadratic (ax^2 + bx + c = 0)")
    print("2. Cubic (ax^3 + bx^2 + cx + d = 0)")
    print("3. Quartic (ax^4 + bx^3 + cx^2 + dx + e = 0)")
    
    choice = input("Enter your choice (1/2/3): ")
    
    if choice == "1":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        root1, root2 = find_quadratic_roots(a, b, c)
        print(f"The roots are: {root1} and {root2}")
    
    elif choice == "2":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        roots = find_cubic_roots(a, b, c, d)
        print("The roots are:", roots)
    
    elif choice == "3":
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        e = float(input("Enter coefficient e: "))
        roots = find_quartic_roots(a, b, c, d, e)
        print("The roots are:", roots)
    
    else:
        print("Invalid choice. Please select 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
