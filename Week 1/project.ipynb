# Financial Calculations Assignment

This notebook calculates:

1. Simple Interest
2. Compound Interest
3. Annuity Plan

```python
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
T = float(input("Enter Time: "))

A = P * (1 + (R/100) * T)

print("Simple Interest Amount:", A)
```

```python
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
n = float(input("Enter number of compounding periods per year: "))
t = float(input("Enter time in years: "))

A = P * (1 + (R/(100*n))) ** (n*t)

print("Compound Interest Amount:", A)
```

```python
PMT = float(input("Enter periodic payment: "))
R = float(input("Enter annual interest rate: "))
n = float(input("Enter compounding periods per year: "))
t = float(input("Enter time in years: "))

rate = R/(100*n)

A = PMT * (((1 + rate) ** (n*t) - 1) / rate)

print("Future value of annuity:", A)
```
